```python
# 1. Load cars data as dataframe using pandas and create a bar plot between
# number of cylinders and frequency of cars with that many number of
# cylinders
# ● Set xlabel as Number of cylinders
# ● Set ylabel as Frequency of cars
# ● Draw a bar plot

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cars-1.csv")
df.head()
```

```python
# Create a bar plot
cylinder_counts = df['cyl'].value_counts().sort_index()
cylinder_counts.plot(kind='bar')

# Set labels and title
plt.xlabel('Number of cylinders')
plt.ylabel('Frequency of cars')
plt.title('Frequency of Cars by Number of Cylinders')

# Show the plot
plt.show()
```

```python
# 2. Write code to load data from cars and print a bar graph of count of columns
# with null values

df.isnull().sum().plot(kind='bar')

# Set labels and title
plt.xlabel('Columns')
plt.ylabel('Count of Null Values')
plt.title('Count of Null Values in Columns')

# Show the plot
plt.show()
```

```python
# 3. Use the 'mpg' (Miles Per Gallon column) and draw a histogram:
# ● Set xlabel: Miles per gallon
# ● Set ylabel: Frequency
# ● Set title as Miles Per Gallon Histogram
# ● Use mpg column to generate a histogram

# Create a histogram for 'mpg' column
plt.hist(df['mpg'], bins=10, edgecolor='black')

# Set labels and title
plt.xlabel('Miles per gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')

# Show the plot
plt.show()
```

```python
# 4. Draw a boxplot on the card dataframes hp column:
# ● Set xlabel: Car Horsepower
# ● Set title as Boxplot for car horsepower
# ● Use hp column to generate a boxplot

# Create a boxplot for 'hp' column
plt.boxplot(df['hp'],patch_artist=True)

# Set labels and title
plt.xlabel('Car Horsepower')
plt.title('Boxplot for Car Horsepower')

# Show the plot
plt.show()

```
