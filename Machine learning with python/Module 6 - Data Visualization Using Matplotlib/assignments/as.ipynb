```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cars-1.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['model'],df['hp'],marker='p',color='b')
plt.xlabel("Models of cars")
plt.ylabel('Hourse-Power of cars')
plt.title('Model Names Vs HoursePower')
# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.bar(df['carb'],df['gear'])
plt.xlabel("Number of craburetors")
plt.ylabel("Number of forward gears")
plt.title("carbs vs gear")
plt.show()
```

```python
# assignment 3

plt.hist(df['wt'],bins=35)
plt.xlabel("weight of the cars")
plt.ylabel("Count")
plt.title("Histogram fr the weight values")
plt.show()
```

```python
# assignment 4

plt.pie(df["cyl"].value_counts(),labels=df['cyl'].unique() ,autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Cylinders')
plt.show()
```

```python
plt.pie(df["model"].value_counts(),labels=df['model'].unique() ,autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Models')
plt.show()
```

```python
#assignment 5

# Set up the area chart
# plt.figure(figsize=(10, 6))
plt.stackplot(df['am'], df['carb'])

# Set labels and title
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs Number of Carburetors')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```
