```python
# 1. Building the k-means clustering algorithm:
# a. Start off by extracting the ‘mpg’, ‘disp’ & ‘hp’ columns from the ‘mtcars’ data.frame.
# Store the result in ‘car_features’

import pandas as pd

# Load the 'mtcars' dataset
mtcars = pd.read_csv('cars-1.csv')

# Extract the 'mpg', 'disp' and 'hp' columns and store in 'car_features'
car_features = mtcars[['mpg', 'disp', 'hp']]
```

```python
#b. Build the kmeans algorithm on top of ‘car_features’. Here, the number of clusters
# should be 3

from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=42)
km.fit(car_features)
```

```python
# c. Bind the clustering vector to ‘car_features’.

cluster_labels = km.labels_

car_features['cluster'] = cluster_labels
```

```python
# d. Extract observations belonging to individual clusters

cluster0 = car_features[car_features['cluster']==0]
cluster1 = car_features[car_features['cluster']==1]
cluster2 = car_features[car_features['cluster']==2]
```
