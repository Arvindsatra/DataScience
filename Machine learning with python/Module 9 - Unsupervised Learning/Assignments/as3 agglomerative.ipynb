```python
# 1. Load iris data from load_iris function from sklearn.datasets package

from sklearn.datasets import load_iris 

iris = load_iris()

data = iris.data
target = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Print some information about the dataset
print("Feature Names:", feature_names)
print("Target Names:", target_names)
print("First 5 Data Samples:\n", data[:5])
print("First 5 Target Labels:", target[:5])
```

```python
# 3. Train an Agglomerative Clustering model based on the data
# 4. Plot dendrogram to visualize the clustering linkage

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

#  Step 3: Train an Agglomerative Clustering model
n_clusters = 3
agglomerative = AgglomerativeClustering(n_clusters=n_clusters)
cluster_labels = agglomerative.fit_predict(data)

# Step 4: Plot dendrogram
linkage_matrix = linkage(data,method='ward')
plt.figure(figsize=(10,7))
dendrogram(linkage_matrix,labels=cluster_labels,leaf_rotation=90)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```
