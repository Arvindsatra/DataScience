```python
data=df.drop(['Unnamed: 1','Unnamed: 2', 'Unnamed: 5'],axis=1)
data.dropna(inplace=True)
data['#Transactions in which\n the item is bought']=data['#Transactions in which\n the item is bought'].astype(int)
data.head()
```

```python

data['Transactions from the Database']
```

```python
# 3. Using the Apriori algorithm generate a list of items frequently bought
# together

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Convert the 'Transactions from the Database' column to a list of sets
data['Transactions from the Database'] = data['Transactions from the Database'].apply(lambda x: x[x.find('{')+1:x.find('}')] if isinstance(x, str) else '')
data
```

```python
# Create a DataFrame
data_df = pd.DataFrame(data)

# Create a list to store the repeated values
repeated_values = []

# Iterate through each row in the DataFrame
for _, row in data_df.iterrows():
    repeated_values.extend([row["Transactions from the Database"]] * row["#Transactions in which\n the item is bought"])

# Generate a DataFrame with the repeated values
final_df = pd.DataFrame({'Transactions from the Database': repeated_values})
final_df.head()
```

```python
# Assuming 'column_name' is the name of the column containing the strings
separator = ', '  # Adjust the separator based on your data
list_of_lists = [row.split(separator) for row in final_df['Transactions from the Database']]

```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(final_df).transform(final_df)
final_df = pd.DataFrame(te_ary, columns=te.columns_)
final_df
```

```python
# Apply the Apriori algorithm to find frequent itemsets
min_support = 0.1  # Adjust the minimum support as needed
frequent_itemsets = apriori(final_df, min_support=min_support, use_colnames=True)

# Display the frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)

```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Generate association rules
association_rules_result = association_rules(frequent_itemsets, metric="support")

# Display the generated association rules
print("Generated Association Rules:")
print(association_rules_result)
```
