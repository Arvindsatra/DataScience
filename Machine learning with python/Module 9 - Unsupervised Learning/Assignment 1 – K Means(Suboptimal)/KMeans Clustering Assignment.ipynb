```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Load customer data.

df = pd.read_csv('customers.csv')
```

```python
# Check the number of cells in each column with null values.
null_counts = df.isnull().sum()
print(null_counts)
```

```python
#Create a scatter plot with Age as X and Spending Score as Y.


plt.scatter(df['Age'], df['Spending Score (1-100)'])

plt.xlabel('Age')
plt.ylabel('spending score')
plt.title('Age vs spending score')
plt.show()
```

```python
#Draw a scatter plot displaying data points colored on the basis of clusters.

X = df[['Age', 'Spending Score (1-100)']]
km = KMeans(n_clusters=4)

# predict the cluster labels
y_pred = km.fit_predict(X)
X['cluster']=y_pred
X0 = X[X['cluster']==0]
X1 = X[X['cluster']==1]
X2 = X[X['cluster']==2]
X3 = X[X['cluster']==3]

# plot the data points colored on the basis of clusters
plt.scatter(X0['Age'], X0['Spending Score (1-100)'], color="yellow")
plt.scatter(X1['Age'], X1['Spending Score (1-100)'], color="red")
plt.scatter(X2['Age'], X2['Spending Score (1-100)'], color="blue")
plt.scatter(X3['Age'], X3['Spending Score (1-100)'], color="green")
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Scatter plot with clusters')
plt.show()
```
