```python
# Perform the following tasks for answering the remaining questions
# ● Rename the last column as Annual Income
census.rename(columns={'annual_income':'Annual Income'},inplace=True)

# ● Remove the missing values from the dataset
census.dropna(inplace=True)

# ● Change the labels of categorical data into numerical data using Label
# Encoder.
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
categorical_columns = census.select_dtypes(include=['object']).columns
for columns in categorical_columns:
    census[columns] = label.fit_transform(census[columns])

# ● Split the dataset into a train and test of proportions 70:30 and set the random
# state to 0.
from sklearn.model_selection import train_test_split

x = census.drop('Annual Income',axis=1)
y = census['Annual Income']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=44)

# ● Build a Logistic Regression Model on the data.
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(x_train,y_train)
y_pred = log.predict(x_test)


# Answer the following questions with the help of the above-created model.
```

```python
# 14.What is the accuracy score of the above model?
# a. 0.60 to 0.70
# b. 0.40 to 0.60
# c. 0.70 to 0.85
# d. None of the above
from sklearn.metrics import accuracy_score, confusion_matrix

print(f'{accuracy_score(y_test,y_pred):.2f}')
print('c. 0.70 to 0.85')
```

Sensitivity (Recall): How well a method detects positive cases.

Specificity: How well a method identifies negative cases.

Precision: How accurate positive predictions are.

Recall (Sensitivity): How well a method detects positive cases.

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```

```python
# 15.What is the specificity of the above model?
# a. 0.20 to 0.30
# b. 0.30 to 0.40
# c. 0.50 to 0.60
# d. None of the above

#Specificity = TN / (TN + FP)

cm = confusion_matrix(y_test,y_pred)

TN = cm[0, 1]
FP = cm[1, 0]
specificity = TN / (TN + FP)
print(specificity)
print('a. 0.20 to 0.30')
```

```python
# 16.What is the model’s precision when the target is False?
# a. 0.60 to 0.70
# b. 0.40 to 0.60
# c. 0.70 to 0.80
# d. None of the above

#Precision (Negative class) = TN / (TN + FN)

print(report)
print('c. 0.70 to 0.80')
```

F1 Score: The F1 score is a metric that combines both precision and recall into a single value. It's particularly useful when the class distribution is imbalanced. The F1 score takes into account both false positives and false negatives, providing a balanced measure of a model's accuracy. It's calculated as the harmonic mean of precision and recall:

F1 Score = 2 * (Precision * Recall) / (Precision + Recall)
