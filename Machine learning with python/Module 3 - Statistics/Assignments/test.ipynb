```python
# 6) Calculate the 95% confidence interval of the given sample using the t-distribution. Sample = [78 65 20 69 36 81 85 71 44 71 47 37 89 25 73 55 80 52 46 77].
# (54.500 , 68.649)
# (50.400 , 69.699)
# (40.500 , 63.659)
# (45.450 , 67.599)

import numpy as np
from scipy.stats import t

# Given sample data
sample_data = np.array([78, 65, 20, 69, 36, 81, 85, 71, 44, 71, 47, 37, 89, 25, 73, 55, 80, 52, 46, 77])

# Parameters
confidence_level = 0.95
degrees_of_freedom = len(sample_data) - 1
mean = np.mean(sample_data)
std_dev = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation

# Calculate the standard error
standard_error = std_dev / np.sqrt(len(sample_data))

# Calculate the t-value for the given confidence level and degrees of freedom
t_value = t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

# Calculate the margin of error
margin_of_error = t_value * standard_error

# Calculate the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

confidence_interval = (lower_bound, upper_bound)
print(confidence_interval)
```

```python
# 20) There are two machines for spray coating at desired thickness. Their variability should be similar if they need to be used for the same applications. To test this, various samples were collected from both machines and the sample covariance of machine 1 and 2 are 1.84 and 6.36 respectively. Comment on whether these machines can be used for the same application if the machines were truly built to have the same variance. Perform the test at 95% significance given n1 = 24 n2 = 25
# Accept null hypothesis, variance of both the machines are different
# Reject null hypothesis, variance of both the machines are same
# Reject null hypothesis, variance of both the machines are different
# Accept null hypothesis, variance of both the machines are same


# f-test
from random import sample
from scipy.stats import f

m1 = 1.84
m2 = 6.36
n1 = 24
n2 = 25

f_value  = m1/m2
p = 1 - f.cdf(f_value, n1-1, n2-1)
if p<0.05:
    print('Reject null hypothesis, variance of both the machines are different')
else:
    print('Accept null hypothesis, variance of both the machines are same')
```

```python
# 21) In the contact process for the production of sulphuric acid, temperature of not more than 450°C is desired. The reactor temperature was measured 24 times and it was found that the mean temperature was 450.3°C and the variance was 1.0. Find whether the reactor is operating at the desired condition? Perform the test at 95% significance.
# Accept null hypothesis, reactor operating at the desired condition
# Reject null hypothesis, reactor not operating at the desired condition
# Accept null hypothesis, reactor not operating at the desired condition
# Reject null hypothesis, reactor operating at the desired condition

import numpy as np
import numpy as np
from scipy.stats import ttest_1samp

# Given information
population_mean = 450  # Population mean under null hypothesis
sample_mean = 450.3  # Sample mean
sample_size = 24  # Sample size
sample_variance = 1.0  # Sample variance

# Create synthetic sample data based on the provided mean and variance
synthetic_sample = np.random.normal(loc=sample_mean, scale=np.sqrt(sample_variance), size=sample_size)

# Perform one-sample t-test
t_statistic, p_value = ttest_1samp(synthetic_sample, population_mean)

# Significance level (alpha)
alpha = 0.05  # 95% significance level

# Compare the p-value with the significance level
if p_value <= alpha:
    # Reject the null hypothesis
    print("Reject null hypothesis: Reactor not operating at the desired condition")
else:
    # Accept the null hypothesis
    print("Accept null hypothesis: Reactor operating at the desired condition")

print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
