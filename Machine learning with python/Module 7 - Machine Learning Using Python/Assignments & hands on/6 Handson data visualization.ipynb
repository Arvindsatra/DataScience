```python
# Step 4: Import the required packages and read data from time_seriest.csv in a
# DataFrame

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('time_series.csv')
df.head()
```

```python
# Step 5: Plot a line graph and take a look at Googles historical data about its stock

plt.figure(figsize=(21,14))
plt.plot(df['Date'], df['GOOG'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Step 6: Read data from national parks and take a look at first 5 rows

df = pd.read_csv('national_parks.csv')

df.head()
```

```python
# Step 7: Plot a histogram based on the GrandCanyon column

plt.hist(df['GrandCanyon'])
plt.show()
```

```python
# Step 8: Read data from types_movies.csv and take a look at first 5 rows

df = pd.read_csv('types_movies.csv')
df.head()
```

```python
# Step 9: Plot a pie chart based on percentage of movies and set labels to be
# sector column
plt.pie(df['Percentage'],labels=df['Sector'],autopct='%1.1f%%')
plt.show()
```

```python
# Step 10: Create and visualize a correlation matrix on time_series data using
# heatmaps

df = pd.read_csv('time_series.csv')
matrix = df.corr()
sns.heatmap(matrix,annot=True)
```
