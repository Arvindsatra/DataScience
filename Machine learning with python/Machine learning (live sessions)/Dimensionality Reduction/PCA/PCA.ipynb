```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.drop("Id",axis=1,inplace=True)
```

```python
df.head()
```

```python
y = df.pop("Species")
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(df)
```

```python
x
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca = PCA(random_state =11 )
```

```python
pca.fit(x)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))#scree
```

```python
pca = PCA(random_state =11 ,n_components=2)
```

```python
a = pca.fit_transform(x)
```

```python
a
```

```python
df2 = pd.DataFrame(a,columns=['Pc1','Pc2'])
```

```python
df2.head()
```

```python
df2['label'] = y
```

```python
df2.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='Pc1',y='Pc2',hue='label',data=df2)
```

```python
sns.heatmap(df2.corr(),annot=True)
```
